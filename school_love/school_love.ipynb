```python
import pandas as pd
import random

random.seed(42)

rows = []
colors = ["red", "pink", "white", "blue"]

for _ in range(1000):
    price = random.randint(100, 600)
    has_chocolate = random.randint(0, 1)
    has_card = random.randint(0, 1)
    color = random.choice(colors)
    delivery_time = random.randint(1, 5)

    liked = 0
    if has_chocolate and has_card and price < 400:
        liked = 1
    if color in ["red", "pink"] and delivery_time <= 2:
        liked = 1
    if random.random() < 0.1:  # шум
        liked = 1 - liked

    rows.append([
        price,
        has_chocolate,
        has_card,
        color,
        delivery_time,
        liked
    ])

df = pd.DataFrame(rows, columns=[
    "price",
    "has_chocolate",
    "has_card",
    "color",
    "delivery_time",
    "liked"
])

df.head()
```
